```python
import numpy as np
```

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot([0,x[0]], [0,x[1]], 'o-')
plt.plot([0,y[0]], [0,y[1]], 'o-')
plt.plot([0,z[0]], [0,z[1]], 'o-')
plt.grid()
```

```python
a = x + y + z

plt.figure()
plt.plot([0,a[0]], [0,a[1]], 'o-')
plt.grid()
```

```python
angle_x_y = np.arccos(np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y)))
print('Угол между векторами x, y:',round(np.degrees(angle_x_y)), 'градусов')
```
